```python
import polars as pl
import boto3
import io

s3 = boto3.client(
    "s3",
    aws_access_key_id="test",
    aws_secret_access_key="test",
    endpoint_url="http://172.18.0.6:4566"
)

bucket_name = "my-bucket"
prefix = "tables/btc_prices/"

# Lista los archivos Parquet y los lee en un solo paso
files = [
    obj["Key"] for obj in s3.list_objects_v2(Bucket=bucket_name, Prefix=prefix).get("Contents", [])
    if obj["Key"].endswith(".parquet")
]

df = pl.concat([
    pl.read_parquet(io.BytesIO(s3.get_object(Bucket=bucket_name, Key=file)["Body"].read()))
    for file in files
])

print(df.head())
```

## Cálculo de las medias

```python
import polars as pl

# Supongamos que 'df' es tu DataFrame de Polars con las columnas necesarias y que lo ordenas por "Date"
df = df.sort("Date")

# Calcular las medias móviles:
# - SMA_14: ventana de 14 filas (incluye la fila actual y 13 anteriores)
# - SMA_50: ventana de 50 filas
# - SMA_200: ventana de 200 filas
df = df.with_columns([
    pl.col("Close").rolling_mean(window_size=14, min_samples=14).alias("SMA_14"),
    pl.col("Close").rolling_mean(window_size=50, min_samples=50).alias("SMA_50"),
    pl.col("Close").rolling_mean(window_size=200, min_samples=200).alias("SMA_200"),
])

# Seleccionar y mostrar las columnas de interés
print(df.select(["Date", "Close", "High", "Low", "Open", "Volume", "SMA_14", "SMA_50", "SMA_200"]).head(5))
```

## Cálculo del RSI

```python
import polars as pl

# Supongamos que 'df' es tu DataFrame y ya está ordenado por "Date"
df = df.sort("Date")

# Paso 1: Calcular el cambio diario y crear "Change"
df = df.with_columns((pl.col("Close") - pl.col("Close").shift(1)).alias("Change"))

# Paso 2: Calcular "Gain" y "Loss" en pasos separados
df = df.with_columns([
    pl.when(pl.col("Change") > 0).then(pl.col("Change")).otherwise(0).alias("Gain"),
    pl.when(pl.col("Change") < 0).then(-pl.col("Change")).otherwise(0).alias("Loss")
])

# Paso 3: Calcular las medias móviles de 14 días para "Gain" y "Loss"
df = df.with_columns([
    pl.col("Gain").rolling_mean(window_size=14, min_samples=14).alias("Avg_Gain"),
    pl.col("Loss").rolling_mean(window_size=14, min_samples=14).alias("Avg_Loss")
])

# Paso 4: Calcular RS y RSI en una primera llamada a with_columns
df = df.with_columns([
    (pl.col("Avg_Gain") / pl.col("Avg_Loss")).alias("RS"),
    (100 - (100 / (1 + (pl.col("Avg_Gain") / pl.col("Avg_Loss"))))).alias("RSI")
])

# Paso 5: Ahora, usar "RSI" para definir las señales de compra y venta
df = df.with_columns([
    pl.when(pl.col("RSI") < 30).then(1).otherwise(0).alias("Buy_Signal"),
    pl.when(pl.col("RSI") > 70).then(1).otherwise(0).alias("Sell_Signal")
])

# Mostrar resultados (omitiendo filas con valores nulos, por ejemplo, al inicio)
result = df.select(["RSI", "RS"]).drop_nulls()
print(result.head(5))
```

## Cálculo de los breakouts

```python
import polars as pl

# Aseguramos que el DataFrame esté ordenado por "Date"
df = df.sort("Date")

# Calcular el máximo de "High" y el mínimo de "Low" de los 20 días previos (excluyendo el día actual)
df = df.with_columns([
    pl.col("High")
      .shift(1)  # Desplaza la columna para excluir el día actual
      .rolling_max(window_size=20, min_samples=20)
      .alias("Max_High"),
    pl.col("Low")
      .shift(1)
      .rolling_min(window_size=20, min_samples=20)
      .alias("Min_Low")
])

# Generar señales de breakout:
# - Breakout_Up: 1 si el precio de cierre supera el máximo de los 20 días previos
# - Breakout_Down: 1 si el precio de cierre cae por debajo del mínimo de los 20 días previos
df = df.with_columns([
    pl.when(pl.col("Close") > pl.col("Max_High")).then(1).otherwise(0).alias("Breakout_Up"),
    pl.when(pl.col("Close") < pl.col("Min_Low")).then(1).otherwise(0).alias("Breakout_Down")
])

# Seleccionar las columnas de interés y mostrar los primeros 10 registros
result = df.select([
    "Date", "Close", "High", "Low",
    "Max_High", "Min_Low",
    "Breakout_Up", "Breakout_Down"
])
print(result.head(10))
```

```python
import polars as pl
import boto3
import io

s3 = boto3.client(
    "s3",
    aws_access_key_id="test",
    aws_secret_access_key="test",
    endpoint_url="http://172.18.0.6:4566"
)

bucket_name = "my-bucket"
prefix = "tables/btc_processed/"

# Lista los archivos Parquet y los lee en un solo paso
files = [
    obj["Key"] for obj in s3.list_objects_v2(Bucket=bucket_name, Prefix=prefix).get("Contents", [])
    if obj["Key"].endswith(".parquet")
]

df = pl.concat([
    pl.read_parquet(io.BytesIO(s3.get_object(Bucket=bucket_name, Key=file)["Body"].read()))
    for file in files
])

print(df.head())
```

## SMA Crossover

```python
import polars as pl

# Asegúrate de que el DataFrame esté ordenado por "Date"
df = df.sort("Date")

# Paso 1: Crear la columna "SMA_14_Previa" desplazando "SMA_14" en 1
df = df.with_columns(
    pl.col("SMA_14").shift(1).alias("SMA_14_Previa")
)

# Paso 2: Definir señales de compra y venta basadas en la comparación de SMA_14 y SMA_50
df = df.with_columns([
    pl.when((pl.col("SMA_14_Previa") < pl.col("SMA_50")) & (pl.col("SMA_14") > pl.col("SMA_50")))
      .then(1).otherwise(0)
      .alias("Buy_Signal"),
    pl.when((pl.col("SMA_14_Previa") > pl.col("SMA_50")) & (pl.col("SMA_14") < pl.col("SMA_50")))
      .then(1).otherwise(0)
      .alias("Sell_Signal")
])

# Paso 3: Filtrar y mostrar las filas donde haya señales de compra o venta
result = df.filter(
    (pl.col("Buy_Signal") == 1) | (pl.col("Sell_Signal") == 1)
).select(["Date", "Close", "SMA_14", "SMA_50", "Buy_Signal", "Sell_Signal"])

print(result.head(10))
```

```python
import polars as pl
import pandas as pd
import time

def simulate_sma_crossover(group_df: pl.DataFrame) -> pl.DataFrame:
    # Convertir el grupo a Pandas para iterar secuencialmente, ya que Polars está optimizado para operaciones
    # vectorizadas, y esto es un proceso secuencial y stateful, 
    # en el que el resultado de cada fila depende del estado calculado en filas anteriores (por ejemplo, el capital o los BTC actuales).
    pdf = group_df.to_pandas()
    pdf["Date"] = pd.to_datetime(pdf["Date"])
    pdf = pdf.sort_values("Date").reset_index(drop=True)
    
    # Inicializar variables de estado
    capital = 10000.0  # Capital inicial en USD
    btc_holdings = 0.0  # BTC en cartera
    
    # Listas para almacenar resultados
    capital_list = []
    btc_list = []
    port_list = []
    
    # Iterar sobre cada fila en orden
    for _, row in pdf.iterrows():
        close = row["Close"]
        buy_signal = row["Buy_Signal"]
        sell_signal = row["Sell_Signal"]
        
        # Señal de compra: si hay señal y no tenemos BTC, compramos con todo el capital
        if buy_signal == 1 and btc_holdings == 0:
            btc_holdings = capital / close
            capital = 0.0
        
        # Señal de venta: si hay señal y tenemos BTC, vendemos todo lo que tenemos
        if sell_signal == 1 and btc_holdings > 0:
            capital = btc_holdings * close
            btc_holdings = 0.0
        
        # Valor total de la cartera
        portfolio_value = capital + (btc_holdings * close)
        
        capital_list.append(capital)
        btc_list.append(btc_holdings)
        port_list.append(portfolio_value)
    
    # Agregar las nuevas columnas
    pdf["Capital"] = capital_list
    pdf["BTC_Holdings"] = btc_list
    pdf["Portfolio_Value"] = port_list
    
    # Seleccionar las columnas definidas en el esquema
    pdf = pdf[["Date", "Close", "Buy_Signal", "Sell_Signal", "Capital", "BTC_Holdings", "Portfolio_Value"]]
    
    # Convertir nuevamente a Polars
    return pl.from_pandas(pdf)

# Aplicamos la función a cada grupo usando groupby.apply y eliminamos la columna de grupo.
start = time.perf_counter()  # Puedes usar time.time() o time.perf_counter()
result_df = simulate_sma_crossover(df)
end = time.perf_counter()

# Ordenamos el resultado por fecha y mostramos las primeras filas
result_df = result_df.sort("Date")
print(f"Tiempo de ejecución con Polars: {end - start:.4f} segundos")
print(result_df)
```

```python
# Ordenar el DataFrame por "Date" en orden descendente y obtener el primer valor de "Portfolio_Value"
final_value = result_df.sort("Date", descending=True)["Portfolio_Value"][0]

initial_capital = 10000.0
returns = (final_value - initial_capital) / initial_capital * 100

print(f"📊 Retorno final: ${final_value:.2f}")
print(f"📈 Retorno relativo: {returns:.2f}%")
```

```python
import polars as pl

import numpy as np
import polars as pl

# Convertir la columna "Portfolio_Value" a un arreglo de NumPy
portfolio_values = result_df["Portfolio_Value"].to_numpy()

# Calcular el máximo acumulado usando np.maximum.accumulate
cummax_values = np.maximum.accumulate(portfolio_values)

# Agregar la columna "Max_Portfolio" al DataFrame
result_df = result_df.with_columns(pl.Series("Max_Portfolio", cummax_values))

# Calcular el drawdown usando las columnas resultantes
result_df = result_df.with_columns(
    ((pl.col("Max_Portfolio") - pl.col("Portfolio_Value")) / pl.col("Max_Portfolio")).alias("Drawdown")
)

# Extraer el valor máximo de la columna "Drawdown"
max_drawdown_value = result_df["Drawdown"].max()

# Imprimir el máximo drawdown en porcentaje
print(f"📉 Máximo Drawdown: {max_drawdown_value * 100:.2f}%")
```
